```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("anime-dataset-2023.csv")

# Display the first few rows to understand the structure
print("Original Data:")
print(df.head())

# Identify numeric columns where we want to replace "UNKNOWN" with the average
numeric_columns = ["Score", "Rank", "Popularity", "Members"]

# Replace "UNKNOWN" with NaN for numeric columns to calculate the mean
df[numeric_columns] = df[numeric_columns].replace("UNKNOWN", np.nan)

# Calculate the mean for each numeric column and fill NaN values
for column in numeric_columns:
    mean_value = df[column].astype(float).mean()  # Convert to float to calculate mean
    df[column].fillna(mean_value, inplace=True)  # Replace NaN with the mean value

# Display the processed data
print("\nProcessed Data:")
print(df.head())

# Optionally, save the processed DataFrame to a new CSV file
df.to_csv("processed_anime_dataset.csv", index=False)
```
